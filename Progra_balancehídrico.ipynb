### Fundamento teórico
* Determinación de la evapotranspiración potencial (**ETP**) mensual según Holdridge:

$$ETP_{mensual} = \frac{(58.93)\,(N°\,dia\,del\,mes)\,(T°media_{mensual})}{N°\,dias\,año}$$ 

* Determinación del balance hídrico (**BH**):

$${Balance\,hídrico} = {Precipitación} - {ETP_mensual}$$

* Determinación de la evapotranspiración real (**ETR**) bajo estas condiciones:

     * $BH_{mes\,actual}>0$ , entonces $ETP =ETR$<br>
     * $BH_{mes\,actual}<0$, entonces verificamos su BH en el mes anterior<br> 
     * $BH_{mes\,anterior}>0 $, entonces $ETR = BH + Precipitación$, con la condición que esta suma sea $<=$ al ETP actual <br> 
     * $BH_{mes\,anterior}>0 $, entonces $ETR = Precipitación$<br>
